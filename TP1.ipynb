```python
# Importation des librairies nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

# Configuration esthétique pour les plots
sns.set(style="whitegrid")

# Chargement des données
housing = fetch_california_housing(as_frame=True)
df = housing.frame

# Renommer la variable cible pour plus de clarté 
df.rename(columns={'MedHouseVal': 'Target'}, inplace=True)

# Affichage des infos
print("Description du DataFrame:")
print(df.info())
print("\nPremières lignes:")
display(df.head())
```

### 1.2 Data summarization and wrangling

```python
# Tâche 1.2: Statistiques descriptives
display(df.describe())

# Vérification des valeurs manquantes
missing_values = df.isnull().sum()
print("\nValeurs manquantes par colonne :")
print(missing_values)

```

Plan de gestion : Si des valeurs manquaient, nous pourrions utiliser 
 - l'imputation (SimpleImputer de sklearn)en remplaçant les NaN par la médiane ou la moyenne de la colonne.
 - la suppression de ces valeurs manquantes si elles ne sont pas nombreuses.

```python
# Tâche 1.3: Séparation Features (X) et Target (y)
X = df.drop('Target', axis=1)
y = df['Target']
```

En observant la matrice, MedInc (Revenu Médian) a une forte corrélation positive (environ 0.69) avec Target. Cela signifie que plus le revenu médian dans un quartier est élevé, plus le prix des maisons tend à être élevé.

## PHASE 2 : FEATURE ENGINEERING AND SELECTION

### 2.1 Feature extraction and engineering

```python
# Tâche 2.1: Feature Engineering
# Copie pour éviter les warnings sur le dataframe original
X_eng = X.copy()

X_eng['RoomsPerHousehold'] = X_eng['AveRooms'] / X_eng['AveOccup']
X_eng['BedroomsPerRoom'] = X_eng['AveBedrms'] / X_eng['AveRooms']
X_eng['PopulationPerHousehold'] = X_eng['Population'] / X_eng['AveOccup']

# Division Train/Test [cite: 60]
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_eng, y, test_size=0.2, random_state=42
)
```

```python
# Tâche 2.2: Standardisation
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit sur le TRAIN uniquement, puis transform sur TRAIN et TEST
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Conversion en DataFrame pour garder les noms de colonnes (plus propre)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

print("Données normalisées (5 premières lignes) :")
display(X_train_scaled.head())
```

Question : Pourquoi fit_transform sur train et transform sur test ?  Réponse : C'est crucial pour éviter le Data Leakage (fuite de données). Le modèle ne doit rien "savoir" des données de test (y compris leur moyenne ou leur écart-type) pendant l'entraînement. Si on calculait la moyenne sur tout le dataset, le test set influencerait l'entraînement, ce qui fausserait l'évaluation de la performance réelle.

```python
# Tâche 2.3: Feature Importance avec RandomForest
from sklearn.ensemble import RandomForestRegressor

rf_selector = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_selector.fit(X_train_scaled, y_train)

# Affichage des importances
importances = pd.Series(rf_selector.feature_importances_, index=X_train.columns)
top_5_features = importances.sort_values(ascending=False).head(5)

print("Top 5 des caractéristiques les plus importantes :")
print(top_5_features)
```

## PHASE 3 : BUILDING, TUNING, DEPLOYING MODELS

### 3.1 Model building and evaluation

```python
# Tâche 3.1 & 3.3 : Entraînement et Évaluation Nous entraînons les deux modèles et calculons le RMSE.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Régression Linéaire
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)
y_pred_lin = lin_reg.predict(X_test_scaled)

mse_lin = mean_squared_error(y_test, y_pred_lin)
rmse_lin = np.sqrt(mse_lin)

# 2. Forêt Aléatoire (Modèle plus robuste)
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
forest_reg.fit(X_train_scaled, y_train)
y_pred_forest = forest_reg.predict(X_test_scaled)

mse_forest = mean_squared_error(y_test, y_pred_forest)
rmse_forest = np.sqrt(mse_forest)

print(f"Régression Linéaire RMSE : {rmse_lin:.4f}")
print(f"Forêt Aléatoire RMSE     : {rmse_forest:.4f}")
```

Tâche 3.2 (Théorie) : Avantage du modèle non-linéaire  Réponse : La Régression Linéaire suppose que la relation entre les variables (ex: Revenu) et la cible (Prix) est une ligne droite. Or, dans l'immobilier, les relations sont souvent complexes (ex: saturation des prix, effets de seuil). Le Random Forest est un modèle non-linéaire capable de capturer ces motifs complexes et les interactions entre variables sans qu'on ait besoin de les spécifier manuellement.


Question : Quel est le meilleur modèle ?  Réponse : Le Random Forest aura un RMSE plus faible (donc meilleur), car il s'adapte mieux à la complexité des données géographiques et économiques

### 3.3 Model interpretation 

```python
# Tâche 3.4: Visualisation Feature Importance du modèle final
plt.figure(figsize=(10, 6))
feat_importances = pd.Series(forest_reg.feature_importances_, index=X_train.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('Importance des caractéristiques (Random Forest)')
plt.xlabel('Score d\'importance')
plt.show()

```

Corrélation avec Tâche 2.3 :
Les résultats doivent être identiques ou très proches de l'étape de sélection.
MedInc devrait être la variable dominante.

### 3.4 Model deployment 

```python
import joblib

# Sauvegarde
filename = 'random_forest_regressor.pkl'
joblib.dump(forest_reg, filename)
print(f"Modèle sauvegardé sous : {filename}")

# Test de rechargement pour vérifier que ça marche
loaded_model = joblib.load(filename)
sample_pred = loaded_model.predict(X_test_scaled.iloc[:1])
print(f"Prédiction test (modèle chargé) : {sample_pred[0]:.4f}")
print(f"Vraie valeur : {y_test.iloc[0]:.4f}")
```
